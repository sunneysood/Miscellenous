```
import numpy as np
```

```
men_height = np.random.normal(69.1,2.9,1000)
men_height
```

```
women_height = np.random.normal(63.7,2.7,1000)
women_height
```

```
import matplotlib.pyplot as plt

# plotting the line 1 points 
plt.plot(men_height,label = "Men")
plt.plot(women_height,label = "Women")
plt.legend()
plt.show()
```

```
plt.hist(men_height, bins = 10)
plt.show()
```

```
plt.hist(women_height, bins = 10)
plt.show()
```
